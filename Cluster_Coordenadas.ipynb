```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder
```

Se debe realizar la categorizacion de las filial

```python
codificador = OneHotEncoder()
codificacion = codificador.fit_transform(df[['Filial']])
```

```python
print(codificacion)
```

```python
nuevas_cols = pd.DataFrame(codificacion.toarray(),columns=codificador.categories_)
nuevas_cols
```

Se generan varias columnas, una por filial y asigna unos valores. 
Ahora debemos unir estas nuevas columnas al dataframe original


```python
datos = pd.concat([df['Latitud'],df['Longitud'],nuevas_cols],axis="columns")
datos
```

El numero de cluster se determinan validando el valor de la inercia, entre una indercia sea  <=1.  En nuestro caso y dado que sabemos que hay 6 filiales, y en 2 de ellas hay 3 y 2 zonas el valor de cluster serÃ­a 9, que corresponde a las zonas donde se define piloto

```python
kmeans = KMeans(n_clusters=9).fit(datos.values)
kmeans.labels_
```

```python
print(kmeans.cluster_centers_, kmeans.inertia_)
```

```python
for cluster in range(kmeans.n_clusters):
  Filial = ''
  if kmeans.cluster_centers_[cluster][2] > 0.9  and kmeans.cluster_centers_[cluster][2] < 1.1 :
    Filial = 'AFINIA'
  if kmeans.cluster_centers_[cluster][3] > 0.9  and kmeans.cluster_centers_[cluster][3] < 1.1 :
    Filial = 'CENS'
  if kmeans.cluster_centers_[cluster][4] > 0.9  and kmeans.cluster_centers_[cluster][4] < 1.1 :
    Filial = 'CHEC'
  if kmeans.cluster_centers_[cluster][5] > 0.9  and kmeans.cluster_centers_[cluster][5] < 1.1 :
    Filial = 'EDEQ'
  if kmeans.cluster_centers_[cluster][6] > 0.9  and kmeans.cluster_centers_[cluster][6] < 1.1 :
    Filial = 'EPM'
  if kmeans.cluster_centers_[cluster][7] > 0.9  and kmeans.cluster_centers_[cluster][7] < 1.1 :
    Filial = 'ESSA'
  print(kmeans.cluster_centers_[cluster][0],' - ' ,kmeans.cluster_centers_[cluster][1], ' - ', Filial)

```
